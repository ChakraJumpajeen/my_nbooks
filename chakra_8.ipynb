# Exercise 8: Analyze the avocado data

```python
# 1. Import the pandas module and the seaborn module.
import pandas as pd
import seaborn as sns
```

```python
# 2. Read avocado.csv into a Pandas DataFrame and display the first five rows.
avocadoes = pd.read_csv('avocado.csv')
avocadoes.head(5)
```

```python
# 3. Display basic information about the DataFrame and its columns. Note the Date column has the object type.
avocadoes.info()
```

```python
# 4. Convert the Date column to the datetime type. Repeat #3.
date_cols = ['Date']
#avocadoes[date_cols].apply(pd.to_datetime)
avocadoes['Date'] = pd.to_datetime(avocadoes['Date'])
avocadoes.info()
```

```python
# 5. Filter the data with query() so it contains just the rows for 2015, for the conventional type, and for the region named Albany. 
#    Store only the Date, Total Bags, and Small Bags columns in a new DataFrame. Display the first 5 rows.
avocadoes_filtered = avocadoes.query('year == 2015 & type == "conventional" & region == "Albany"')[['Date', 'Total Bags', 'Small Bags']]
avocadoes_filtered.head(5)
```

```python
# 6. Melt the data in the Total Bags and Small Bags columns, but not the values in the Date column, 
#    naming the column that contains the type of bag as Bags and the column that contains the number of bags as Count. Display the melted DF.

avocadoes_melted = avocadoes.melt(id_vars='Date',
                           value_vars=['Total Bags','Small Bags'], 
                           var_name='Bags', 
                           value_name='Count')

avocadoes_melted
```

```python
# 7. Plot the melted data with a Seaborn line plot, using Date as x-axis, Count as y-axis, and Bags for the hue. 
sns.relplot(data=avocadoes_melted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
# 8. Bin the data in the Total Volume column into four quantiles labeled â€˜Poorâ€™, â€˜Modestâ€™, â€˜Goodâ€™, and â€˜Excellentâ€™, 
#    and store the bin labels in a new column named Sales Volume. Display the first five rows of the modified DF.

avocadoes['Sales Volume'] = pd.qcut(avocadoes['Total Volume'], q=4, labels=['Poor','Modest','Good','Excellent'])
avocadoes.head(5)
```

```python
# 9. Plot the binned data by year with sns.countplot().

sns.countplot(data=avocadoes, x='year', hue='Sales Volume')
```

```python
# 10. Melt the three bag size columns specifying 
#    var_name = Bags and value_name = Count. Display the melted DF.
avocadoes_melted = avocadoes.melt(id_vars='year',
                           value_vars=['Small Bags','Large Bags' ,'XLarge Bags'], 
                           var_name='Bags', 
                           value_name='Count')

avocadoes_melted
```

```python
# 11. Plot the bags by year using sns.relpot() 
#    with kind = line, x = year, y = Count and hue = Bags

sns.relplot(data=avocadoes_melted, kind='line', x='year', y='Count', hue = 'Bags')
```
